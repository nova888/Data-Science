### Initial Data Understanding of Iris Dataset

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset
iris_df = sns.load_dataset('iris')

print("First 5 rows of the Iris dataset:")
display(iris_df.head())
```

```python
# Identify the number of data points (rows) and features (columns)
num_rows, num_cols = iris_df.shape
print(f"Number of data points (rows): {num_rows}")
print(f"Number of features (columns): {num_cols}")
```

```python
# Identify data types and non-null values
print("\nData types and non-null counts:")
iris_df.info()
```

```python
# Identify missing values
print("\nMissing values per column:")
display(iris_df.isnull().sum())
```

```python
# Display descriptive statistics to get a sense of data distribution and potential outliers
print("\nDescriptive statistics of numerical features:")
display(iris_df.describe())
```
