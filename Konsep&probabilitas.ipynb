```python
import numpy as np

np.random.seed(0)
throws = np.random.randint(1, 7, size=10000)
even_count = np.sum(throws % 2 == 0)
probability_even = even_count / len(throws)
print(f"Simulated probability of rolling an even number: {probability_even}")
```

```python
np.random.seed(0)
X = np.random.normal(loc=0, scale=1, size=1000)
Y = np.random.normal(loc=0, scale=1, size=1000)

covarian_matrix = np.cov(X, Y)
covariance = covarian_matrix[0, 1]
print(f"covariance: {covariance}")
```

```python
# Distribusi Uniform

import matplotlib.pyplot as plt

np.random.seed(0)
uniform_data = np.random.uniform(0, 1, 1000)
plt.hist(uniform_data, bins=30, color='lightblue', edgecolor='black')
plt.title('Distribusi Uniform')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Distribusi Binomial

binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(binomial_data, bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribusi Binomial')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

```python
#Distribusi Poisson

poisson_data = np.random.poisson(lam=3, size=1000)
plt.hist(poisson_data, bins=30, color='lightblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
