```python
import pandas as pd


data = {
    "Nama": ["Ani", "Budi", "Citra"],
    "Usia": [25, 30, 35],
    "Kota": ["Jakarta", "Bandung", "Surabaya"]
}
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
data = {'Nama': ['Ani', 'Budi', 'Citra'], 'Usia': [25, 30, 35]}
df = pd.DataFrame(data, index=['A', 'B', 'C'])
print(df.loc['B', 'Usia'])  # Output: 30
```
