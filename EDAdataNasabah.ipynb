```python
import pandas as pd

# Load the dataset with semicolon delimiter
df_nasabah = pd.read_csv('/content/Data_Nasabah.csv', sep=';')
```

### Data Overview: Shape, Information, and Descriptive Statistics

```python
# Display the number of rows and columns
print(f"Number of rows: {df_nasabah.shape[0]}")
print(f"Number of columns: {df_nasabah.shape[1]}\n")

# Display data types and non-null counts
print("Data Types and Non-Null Counts:")
df_nasabah.info()

# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
display(df_nasabah.describe())

# For categorical columns, display unique values and their counts
print("\nUnique Values and Counts for Categorical Columns:")
for column in df_nasabah.select_dtypes(include='object').columns:
    print(f"Column '{column}':")
    display(df_nasabah[column].value_counts())
    print("\n")
```
