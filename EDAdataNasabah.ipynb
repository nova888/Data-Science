```python
import pandas as pd

# Load the dataset with semicolon delimiter
df_nasabah = pd.read_csv('/content/Data_Nasabah.csv', sep=';')
```

### Data Overview: Shape, Information, and Descriptive Statistics

```python
# Display the number of rows and columns
print(f"Number of rows: {df_nasabah.shape[0]}")
print(f"Number of columns: {df_nasabah.shape[1]}\n")

# Display data types and non-null counts
print("Data Types and Non-Null Counts:")
df_nasabah.info()

# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
display(df_nasabah.describe())

# For categorical columns, display unique values and their counts
print("\nUnique Values and Counts for Categorical Columns:")
for column in df_nasabah.select_dtypes(include='object').columns:
    print(f"Column '{column}':")
    display(df_nasabah[column].value_counts())
    print("\n")
```

```python
df_nasabah['jenis_kelamin'] = df_nasabah['jenis_kelamin'].map({'Laki-Laki': 1, 'Perempuan': 2})
df_nasabah['jenis_produk'] = df_nasabah['jenis_produk'].map({'tabungan': 1, 'kartu_kredit': 2, 'deposito': 3})
df_nasabah['pengguna_mobile_banking'] = df_nasabah['pengguna_mobile_banking'].map({'YA': 1, 'TIDAK': 2})
display(df_nasabah.head())
```

### Analisis Distribusi Data Numerik

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns for analysis (including the newly encoded ones)
numerical_cols = ['umur', 'pendapatan', 'saldo_rata_rata', 'jumlah_transaksi',
                  'frekuensi_kunjungi_cabang', 'skor_kredit',
                  'jenis_kelamin', 'jenis_produk', 'pengguna_mobile_banking']

print('Descriptive Statistics for Numerical Columns:')
display(df_nasabah[numerical_cols].describe())

# Plotting histograms for numerical features to visualize their distributions
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1) # Adjust subplot grid based on number of numerical columns
    sns.histplot(df_nasabah[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Matriks Korelasi Antar Variabel Numerik

### Segmentasi Nasabah menggunakan K-Means Clustering

Untuk melakukan segmentasi nasabah, kita akan menggunakan algoritma K-Means Clustering. Langkah-langkahnya adalah:
1.  **Pemilihan Fitur:** Memilih fitur `saldo_rata_rata` dan `jumlah_transaksi`.
2.  **Penskalaan Data:** Menskalakan fitur agar semua memiliki kontribusi yang setara.
3.  **Menentukan Jumlah Kluster (K):** Menggunakan metode Elbow untuk menemukan nilai K yang optimal.
4.  **Penerapan K-Means:** Melakukan klustering dengan nilai K yang telah ditentukan.
5.  **Visualisasi Hasil:** Menampilkan hasil segmentasi dalam scatter plot.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Pemilihan Fitur
X = df_nasabah[['saldo_rata_rata', 'jumlah_transaksi']]

# 2. Penskalaan Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Menentukan Jumlah Kluster (K) menggunakan metode Elbow
wcss = [] # Within-Cluster Sum of Squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Metode Elbow untuk Menentukan Jumlah Kluster Optimal')
plt.xlabel('Jumlah Kluster (K)')
plt.ylabel('WCSS (Inertia)')
plt.grid(True)
plt.show()
```

Dari grafik Elbow method di atas, kita perlu mengidentifikasi 'siku' (elbow) pada grafik untuk menentukan jumlah kluster (K) yang optimal. 'Siku' ini menunjukkan titik di mana penurunan WCSS mulai melambat secara signifikan, yang mengindikasikan bahwa menambahkan lebih banyak kluster setelah titik ini tidak akan memberikan banyak informasi tambahan. Setelah menentukan nilai K yang optimal dari grafik, kita akan melanjutkan dengan penerapan K-Means.
