```python
#Fungsi Linear

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = 2 * x + 1

plt.plot(x, y, label='f(x) = 2x + 1')
plt.title("Linear Function")
plt.legend()
plt.show()
```

```python

#Fungsi Kuadrat


x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 3

plt.plot(x, y, label='f(x) = x^2 - 4x + 3')
plt.title("Quadratic Function")
plt.legend()
plt.show()
```
