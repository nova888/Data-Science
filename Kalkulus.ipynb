```python
#Fungsi Linear

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = 2 * x + 1

plt.plot(x, y, label='f(x) = 2x + 1')
plt.title("Linear Function")
plt.legend()
plt.show()
```

```python

#Fungsi Kuadrat


x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 3

plt.plot(x, y, label='f(x) = x^2 - 4x + 3')
plt.title("Quadratic Function")
plt.legend()
plt.show()
```

```python
#fungsi polinominal

x = np.linspace(-10, 10, 400)
y = 0.1 * x**3 - 0.5 * x**2 + 2 * x - 1

plt.plot(x, y, label='f(x) = 0.1x^3 - 0.5x^2 + 2x - 1')
plt.title("Polynomial Function")
plt.legend()
plt.show()
```

```python
#Fungsi Eksponensial


x = np.linspace(-10, 10, 400)
y = 2**x
plt.plot(x, y, label='f(x) = 2^x')
plt.title("Exponential Function")
plt.legend()
plt.show()
```

```python
x = np.linspace(0.1, 10, 400)
y = np.log(x)
plt.plot(x, y, label='f(x) = log(x)')
plt.title("Logarithmic Function")
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 400)
y = np.sin(x)
plt.plot(x, y, label='f(x) = sin(x)')
plt.title("Trigonometric Function")
plt.legend()
plt.show()
```

```python
def f(x):
    return 3 * x**3 - 6 * x**2 + x - 4

def first_derivative(x):
    return 9 * x**2 - 12 * x + 1

def second_derivative(x):
    return 18 * x - 12

x = np.linspace(-2, 4, 400)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, first_derivative(x), label="f'(x)")
plt.plot(x, second_derivative(x), label="f''(x)")
plt.title("Derivatives of a Polynomial Function")
plt.legend()
plt.show()
```

```python
def f(x):
    return x**2 - 4*x + 3

def df_dx(x):
    return 2*x - 4

def gradient_descent(init_x, learning_rate=0.1, iterations=10):
    x = init_x
    x_history = [x]
    for _ in range(iterations):
        grad = df_dx(x)
        x -= learning_rate * grad
        x_history.append(x)
    return x, x_history

x_min, x_history = gradient_descent(0.0)
x = np.linspace(-2, 4, 400)
y = f(x)

plt.plot(x, y, 'r-', label='f(x) = x^2 - 4x + 3')
plt.plot(x_history, f(np.array(x_history)), 'bo', label='Gradient Descent', markersize=3)
plt.title("Gradient Descent Visualization")
plt.legend()
plt.show()
```

```python
def f(x):
    return x

def F(x):
    return x**2 / 2

x = np.linspace(0, 2, 400)
y = f(x)
F_y = F(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2, label='f(x) = x')
ax.plot(x, F_y, 'g', linewidth=2, label='F(x) = x^2/2')

# Fill area between 0 and 1
ax.fill_between(x, y, where=[(xi >= 0) and (xi <= 1) for xi in x], color='skyblue', alpha=0.5)

ax.set_title("Integral Visualization")
ax.legend()
ax.grid(True)
plt.show()
```
