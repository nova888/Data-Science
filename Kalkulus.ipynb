```python
#Fungsi Linear

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = 2 * x + 1

plt.plot(x, y, label='f(x) = 2x + 1')
plt.title("Linear Function")
plt.legend()
plt.show()
```

```python

#Fungsi Kuadrat


x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 3

plt.plot(x, y, label='f(x) = x^2 - 4x + 3')
plt.title("Quadratic Function")
plt.legend()
plt.show()
```

```python
#fungsi polinominal

x = np.linspace(-10, 10, 400)
y = 0.1 * x**3 - 0.5 * x**2 + 2 * x - 1

plt.plot(x, y, label='f(x) = 0.1x^3 - 0.5x^2 + 2x - 1')
plt.title("Polynomial Function")
plt.legend()
plt.show()
```

```python
#Fungsi Eksponensial


x = np.linspace(-10, 10, 400)
y = 2**x
plt.plot(x, y, label='f(x) = 2^x')
plt.title("Exponential Function")
plt.legend()
plt.show()
```
